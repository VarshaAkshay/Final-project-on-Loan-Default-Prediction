<h4 style="color:Black;">➡️Handling Missing Values</h4>

```python
#Count missing values of each columns
df.isnull().sum()
```

```python
#Filling missing values
# -> Missing values in LoanAmount 
#Median is the best choice as loan amounts can vary a lot.Median gives a fair middle value.
# ->Missing values in CreditScore
# people's scores are not spread evenly(most people are average,only a few have very low or very high).So we use median.
# ->Missing values in Income
# Incomes are usually uneven, only a few people earn a lot more than others.Using median represents a typical income.
# ->Missing values in Age
# Most people are in a normal age(20-60),but if there are a few children or very old .Using median keeps the age correctly.
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")

df["Age"].fillna(df["Age"].median(), inplace=True)
df["Income"].fillna(df["Income"].median(), inplace=True)
df["LoanAmount"].fillna(df["LoanAmount"].median(), inplace=True)
df["CreditScore"].fillna(df["CreditScore"].median(), inplace=True)
df.tail(55)
```

<h5 style="color:Black;"> ✅ Removing Multiple Spaces, Leading /Trailing Spaces(Stripping),Changing to Proper Cases</h5>

```python
#Convert to proper case and fix apostrophes
#categorical_cols = ["Education", "EmploymentType", "MaritalStatus", "LoanPurpose"],We create a list of column names that are categorical (text) columns.These are the columns we want to clean.
#.apply(lambda col: ...).Applies a function to each column in that selection.
# inside the lambda function (applied step by step):
#.astype(str)-Converts all values to string ( sometimes it might contain NaN, numbers, or mixed types).
#str.strip()-Removes leading and trailing spaces.
#.str.replace(r"[^a-zA-Z\s']", "", regex=True)-Removes all characters that are not letters, spaces, or apostrophes.
#.str.replace(r"\s+", " ", regex=True)-Replaces multiple spaces with a single space.
#.str.title()-Converts the string to Title Case (first letter capitalized, rest lowercase).
categorical_cols = ["Education", "EmploymentType", "MaritalStatus", "LoanPurpose"]
df[categorical_cols] = df[categorical_cols].apply(
    lambda col: col.astype(str)
                    .str.strip()                              # remove leading/trailing spaces
                    .str.replace(r"[^a-zA-Z\s']", "", regex=True) # keep only alphabets & '
                    .str.replace(r"\s+", " ", regex=True)     # collapse multiple spaces
                    .str.title()                              # proper case
)
print(df[categorical_cols].tail(15))
```

```python
# Remove hash symbols from LoanPurpose column
df['LoanPurpose'] = df['LoanPurpose'].astype(str).str.replace('#', '', regex=False)

# Check again if hash symbol exists in LoanPurpose column
hash_exists = df['LoanPurpose'].str.contains('#').any()

print("Does '#' still exist in LoanPurpose?:", hash_exists)
```

```python
# Clean 'LoanPurpose' column
df['LoanPurpose'] = df['LoanPurpose'].str.strip()       # remove spaces
df['LoanPurpose'] = df['LoanPurpose'].str.lower()       # make lowercase
```

```python
#Correcting datatypes makes the dataset clean, efficient, and ready for analysis.
# Use of astype()- function is used to change the data type of a column (or entire DataFrame).

# Making a copy of the original  dataframe df.
df_cleaned = df.copy()

# 1. Converting Yes/No columns to boolean of HasDependents,HasCosigner
bool_columns = ["HasDependents", "HasCoSigner"]
for col in bool_columns:
    df_cleaned[col] = df_cleaned[col].map({"Yes": True, "No": False})
    
# 2. Converting Yes/No columns to boolean of HasMortgage
    df_cleaned["HasMortgage"] = df_cleaned["HasMortgage"].astype(bool)


# 3. Converting categorical text columns to category dtype
categorical_columns = ["Education", "EmploymentType", "MaritalStatus", "LoanPurpose"]
for col in categorical_columns:
    df_cleaned[col] = df_cleaned[col].astype("category")

# 4. Converting Default column to boolean (if needed otherwise keep it as integer)
df_cleaned["Default"] = df_cleaned["Default"].astype(bool)

# 5. Converting Age to nullable integer (inplace of float)
df["Age"] = df["Age"].astype("Int64")


#  Now checking the updated datatypes
print(df_cleaned.dtypes)
```

<h6 style="color:Blue;">Loan_to_Income</h6>

```python
# how much of loan amount compared to annual income
#if someone earns ₹10,00,000 annually and takes a loan of ₹5,00,000 → Loan_to_Income = 0.5 (50%).
#  Higher ratio means loan is big compared to income, which is riskier.
df['Loan_to_Income'] = df['LoanAmount'] / df['Income']
```

<h6 style="color:Blue;">Credit_Utilization_Score</h6>

```python
#loan utilization based on credit lines.
#Measures how much loan per credit line the borrower is using.
#+1 is added to avoid dividing by zero when someone has no credit lines.
#If a person has only 1–2 credit lines and high loan amount → their utilization is high, which increases risk.
df['Credit_Utilization'] = df['LoanAmount']/(df['NumCreditLines']+1)
```

<h6 style="color:Blue;">Years_Employed</h6>

```python
#Converted months of employment into years.
#Longer employment history usually indicates stability and lower risk.
df['Years_Employed'] = df['MonthsEmployed']/12
```

<h6 style="color:Blue;">Age_Group</h6>

```python
#Categorized customers(ypung,mid-age,Mature,Senior)
#It creates a new column called AgeGroup in your dataframe by grouping ages into categories (bins).
#pd.cut() → A pandas function used to segment continuous values into ranges (bins).
#bins=[18, 30, 45, 60, 100] → Defines the cut points (intervals) for the bins. ie,18 to 30,30 to 45,45 to 60,60 to 100
#labels=['Young', 'Mid-age', 'Senior', 'Elder'] → Names for these bins.
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 30, 45, 60, 100],
                         labels=['Young', 'Mid-age', 'Mature', 'Senior'])
```

<h6 style="color:Blue;">Risk_Score</h6>

```python
#df['CreditScore']/850,Credit scores are usually between 300–850,Dividing by 850 normalizes the score to a range between 0 and 1,Higher values mean better creditworthiness
#- df['DTIRatio'],Debt-to-Income ratio = debt obligations / income.Higher DTI = more risky, so subtracting this reduces the score.
#- df['Loan_to_Income'],Loan amount relative to income,Larger loan burden compared to income increases risk, so it is subtracted too.
#Custom score: higher= safer,lower=riskier(based on CreditScore,DTI,Loan_to_Income.)
df['Risk_Score'] = (df['Risk_Score']) = (df['CreditScore']/850) - df['DTIRatio'] - df['Loan_to_Income']
```

<h6 style="color:Blue;">Has_Family_Dependents</h6>

```python
#1 if borrower has dependents & not single,else 0
#derived from MaritalStatus and Dependents
#df['MaritalStatus'] != 'Single'.Checks if the borrower’s marital status is not single.Returns True if married/divorced/widowed, etc.Returns False if single.
#df['HasDependents'] == 'Yes',Checks if the borrower has dependents.Returns True if dependents exist, otherwise False.
#& (logical AND)->Combines both conditions:Borrower is not single,Borrower has dependents.Both must be true for the final result to be True.
#.astype(int)->Converts True/False into 1/0.
#Final result stored in new column->A new column HasFamilyDependents is created with:1 if borrower has dependents and is not single,0 otherwise.
df['HasFamilyDependents'] = ((df['MaritalStatus'] != 'Single') & (df['HasDependents'] == 'Yes')).astype(int)
```

<h6 style="color:Blue;">LoanPurpose vs LoanAmount vs Default</h6>

```python
# For each type of loan, what is the average loan amount, and how is it distributed between defaulters and non-defaulters.
# Group the data: LoanPurpose vs LoanAmount grouped by Default
grouped = df.groupby(["LoanPurpose", "Default"])["LoanAmount"].mean().unstack()
# df.groupby(["LoanPurpose", "Default"]) → groups the dataset first by LoanPurpose, then by Default status.
# ["LoanAmount"].mean() → calculates the average LoanAmount for each group
# .unstack() → reshapes the table so that Default values (0 = No, 1 = Yes) become separate columns.
## This gives us a table where rows = LoanPurpose and columns = Default status, values = mean loan amount.
plt.figure(figsize=(10,5), facecolor="lightblue")

# Stacked bar chart
grouped.plot(        # plots the grouped data.
    kind="bar",      # makes a bar chart.
    stacked=True,    # stacks Default=0 and Default=1 values on top of each other inside each LoanPurpose bar
    color=["lightyellow", "lightgreen"],  
    edgecolor="black",
    ax=plt.gca()
)


plt.title("Average Loan Amount by Loan Purpose and Default Status", fontsize=16, weight="bold", color="darkblue")
plt.xlabel("Loan Purpose", fontsize=10, weight="bold")
plt.ylabel("Average Loan Amount", fontsize=10, weight="bold")

plt.legend(    # legend explains colors, line styles
    title="Default (0 = No, 1 = Yes)", # sets a title above the legend 
    loc="upper center", # legend box location
    bbox_to_anchor=(0.5, -0.25), # adjustments done in this legend box, centered hporizontally and pushed slightly below the axis.
    ncol=2) # legend items in 2 columns without stacking.


# observations noted at the side
plt.figtext(1.02,0.5,
            "Observations:\n"
            "•All Loan Purposes Show Defaults.\n"
            "•Business Loans Have the Highest Average Loan Amount.\n"
            "•Education Loans Show Significant Defaults.\n"
            "•Auto Loans Have Lower Average Loan Amounts.\n"
            "•Home Loans Appear More Stable.\n"
            "•Other Category is Mid-Range.\n",
             ha="left", va="center", fontsize=8, weight="bold", color="black")

plt.gca().set_facecolor("lavender")


plt.xticks(color="brown",fontsize=8,rotation=45, ha="right")
plt.yticks(color="brown",fontsize=8)

plt.show()
```
